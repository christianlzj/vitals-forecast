```python
import pandas as pd
from collections import defaultdict

base_path = "../data/vitals_data/mimic_vitals_p0{}.parquet"

patient_records = defaultdict(set)

for i in range(10):  # p00 → p09
    path = base_path.format(i)
    print(f"Processing {path}...")

    df = pd.read_parquet(path, columns=["Patient", "Record"])

    for patient, group in df.groupby("Patient"):
        patient_records[patient].update(group["Record"].unique())

# convert sets to lists if needed
patient_records = {k: list(v) for k, v in patient_records.items()}
```

```python
import pandas as pd
from datetime import datetime


def parse_record_start(record_name: str) -> datetime:
    parts = record_name.split("-")
    year, month, day = parts[1], parts[2], parts[3]
    hour = parts[4]
    minute = parts[5].replace("n", "").replace("p", "")  # strip suffix
    return datetime(int(year), int(month), int(day), int(hour), int(minute))


def subject_id_to_key(subject_id: int) -> str:
    # 79 -> 'p000079'
    return f"p{subject_id:06d}"


def expand_with_records(
    cleaned_path: str,
    patient_records: dict,
    output_path: str,
    chunksize: int = 1_000_000,
):
    # precompute record start times once
    record_starts = {
        record: parse_record_start(record)
        for records in patient_records.values()
        for record in records
    }

    write_header = True

    reader = pd.read_csv(cleaned_path, chunksize=chunksize, low_memory=False)

    for i, chunk in enumerate(reader):
        print(f"Chunk {i + 1}: {len(chunk):,} rows")

        chunk["CHARTTIME"] = pd.to_datetime(chunk["CHARTTIME"])

        expanded_rows = []

        for subj_id, group in chunk.groupby("SUBJECT_ID"):
            key = subject_id_to_key(subj_id)
            records = patient_records.get(key)
            if not records:
                print(f"No records found for {key}")
                continue

            for record in records:
                start_time = record_starts[record]
                expanded = group.copy()
                expanded["RECORD"] = record
                expanded["MINUTE"] = (
                    (expanded["CHARTTIME"] - start_time)
                    .dt.total_seconds()
                    .div(60)
                    .round()
                    .astype(int)
                )
                expanded_rows.append(expanded)

        if not expanded_rows:
            continue

        result = (
            pd.concat(expanded_rows, ignore_index=True)
            .sort_values(["RECORD", "MINUTE"])
            .reset_index(drop=True)
        )

        result.to_csv(
            output_path,
            mode="w" if write_header else "a",
            header=write_header,
            index=False,
        )
        write_header = False
        print(f"  -> {len(result):,} rows written")

        break

    print(f"\nDone. Output written to {output_path}")

expand_with_records(
    cleaned_path="../data/clinical/CLEANED_CHARTEVENTS.csv",
    patient_records=patient_records,
    output_path="../data/clinical/EXPANDED_CHARTEVENTS.csv",
)
```
